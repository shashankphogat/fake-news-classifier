```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
```

```python
data=data.dropna()
```

```python
data=data.reset_index()
```

```python
x=data.title
```

```python
x
```

```python
import re
from nltk.stem import PorterStemmer 
from nltk.corpus import stopwords 
ps = PorterStemmer() 
corpus=[]
for i in range(len(x)):
    intermediate=re.sub('[^a-zA-Z]',' ',x[i])
    intermediate=intermediate.lower()
    intermediate=intermediate.split()
    intermediate=[ps.stem(word) for word in intermediate if not word in stopwords.words('english') ]
    intermediate=' '.join(intermediate)
    corpus.append(intermediate)
corpus
    
```

```python
from tensorflow.keras.preprocessing.text import one_hot
vocab_size=5000
corpus2=corpus.copy()
one_hot_representation=[one_hot(sentence,vocab_size) for sentence in corpus2] 
one_hot_representation
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
sent_length=20
pad_sequenced=pad_sequences(one_hot_representation,padding='pre',maxlen=sent_length)
pad_sequenced
```

```python
#model creation
from tensorflow.keras.layers import Embedding
from tensorflow.keras import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
embedding_features_length=40
model=Sequential()
model.add(Embedding(vocab_size,embedding_features_length,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
y=data.label
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(pad_sequenced, y, test_size=0.33, random_state=42)
```

```python
#model training
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64)
```

```python
#performance metrics
y_pred=model.predict_classes(x_test)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```
